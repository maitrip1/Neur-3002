```python
# Python simulation of lateral inhibition in the horseshoe crab retina
# - 2D grid of ommatidia (each unit = photoreceptor)
# - Direct inhibition between neighbors via lateral nerve fibers
# - No interneurons involved
# - Inhibition occurs peripherally before signal reaches the brain

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1)

def make_stimulus(shape=(40,40), spots=3, spot_amplitude=1.0, spot_radius=3):
    """Create a stimulus image with bright spots on a noisy background."""
    H, W = shape
    stim = 0.05 * np.random.rand(H, W)
    for _ in range(spots):
        cy = np.random.randint(0, H)
        cx = np.random.randint(0, W)
        y, x = np.ogrid[:H, :W]
        mask = (x - cx)**2 + (y - cy)**2 <= spot_radius**2
        stim[mask] += spot_amplitude * (0.8 + 0.4 * np.random.rand())
    return np.clip(stim, 0, None)

def lateral_inhibition_peripheral(stimulus, inhibition_strength=0.65,
                                  neighbor_kernel=None, time_steps=20,
                                  leak=0.1, clamp_min=0.0):
    """
    Iterative model of peripheral lateral inhibition:
      A_i(t+1) = A_i(t) + stimulus_i - inhibition - leak * A_i
    No interneurons  neighbors directly inhibit each other through fiber spread.
    """
    H, W = stimulus.shape

    if neighbor_kernel is None:
        neighbor_kernel = np.array([[0.05, 0.1, 0.05],
                                    [0.1 , 0.0, 0.1 ],
                                    [0.05, 0.1, 0.05]])

    # Normalize kernel (excluding center since center = 0)
    kernel = neighbor_kernel / neighbor_kernel.sum()

    A = stimulus.copy()

    for _ in range(time_steps):
        pad = 1
        Apad = np.pad(A, pad, mode='reflect')
        inh = np.zeros_like(A)
        for dy in range(-1, 2):
            for dx in range(-1, 2):
                ky = dy + 1
                kx = dx + 1
                w = kernel[ky, kx]
                if w != 0:
                    inh += w * Apad[pad+dy:pad+dy+H, pad+dx:pad+dx+W]

        inh_current = inhibition_strength * inh
        A = A + stimulus - inh_current - leak * A
        A = np.clip(A, clamp_min, None)

    return A

# Create basic stimulus
shape = (60, 60)
stim = make_stimulus(shape=shape, spots=4, spot_amplitude=1.6, spot_radius=4)

# Kernels representing narrow vs wide lateral nerve spread
kernel_narrow = np.array([[0.0, 0.3, 0.0],
                          [0.3, 0.0, 0.3],
                          [0.0, 0.3, 0.0]])

kernel_wide = np.array([[0.05, 0.12, 0.05],
                        [0.12, 0.0, 0.12],
                        [0.05, 0.12, 0.05]])

peripheral_narrow = lateral_inhibition_peripheral(
    stim, inhibition_strength=0.8,
    neighbor_kernel=kernel_narrow,
    time_steps=30, leak=0.08)

peripheral_wide = lateral_inhibition_peripheral(
    stim, inhibition_strength=0.7,
    neighbor_kernel=kernel_wide,
    time_steps=30, leak=0.08)

# ---- Visualization ----

plt.figure(figsize=(6,6))
plt.title("Stimulus (photoreceptor drive)")
plt.imshow(stim, interpolation='nearest')
plt.colorbar()
plt.axis('off')

plt.figure(figsize=(6,6))
plt.title("Peripheral output  narrow lateral inhibition")
plt.imshow(peripheral_narrow, interpolation='nearest')
plt.colorbar()
plt.axis('off')

plt.figure(figsize=(6,6))
plt.title("Peripheral output  wide lateral inhibition")
plt.imshow(peripheral_wide, interpolation='nearest')
plt.colorbar()
plt.axis('off')

plt.show()
```
